```python
import pandas as pd
import re
import numpy as np
from datetime import datetime
```

```python
def fill_time(x):
  return str(x)[:-3]+"~"+str(x)[0]+str(int(str(x)[1])+2)+str(x)[2:-3]
```

```python
root_path = "/content/drive/MyDrive/Code/IHCO/" ## 경로 변경 요망
src1 = pd.read_excel(root_path+"dataset/n차 봉사 참가자 명단_배포용.xlsx", sheet_name='정리본').dropna(axis=0)
src1['모임시간'] = src1['모임시간'].apply(fill_time)
src1.rename(columns={'모임시간': '활동시간', '이름': '성함'}, inplace=True)  # 컬럼명 변경
src1.reset_index(inplace=True, drop=True) # index 재설정
```

```python
dst_dict = dict()
columns = ['순번', '활동일자', '활동시간', '성함', '생년월일', '전화번호', '1365 ID', 'VMS ID', '사진 및 동영상 촬영 동의', '개인정보 활용 동의', '체온', '서명']
for f in columns:
  if f not in src1.columns: src1[f] = np.NaN
src1['순번'] = src1.index+1
date = input("n차 봉사 날짜 입력(예: 24.01.01): ")
src1['활동일자'] = datetime.strptime(date,  '%y.%m.%d')
src1 = src1[columns]
```

```python
src2 = pd.read_excel(root_path+"dataset/설문조사_배포용.xlsx")
src2 = src2.loc[:, ['1. 이름 (예시:김보건)(*)',
       '4. 생년월일 (예시: 2000.01.01)(*)', '5. 연락처(*)', '6. 1365 아이디(*)', "7. VMS 아이디(*)"]]
src2_dict = src2.to_dict(orient='list')
```

```python
dst_dict = src1.to_dict(orient='list')
for i, phone in enumerate(dst_dict['전화번호']):
    try:
        idx = src2_dict['5. 연락처(*)'].index(phone)
    except:
        print(phone, dst_dict['성함'][i])
    if dst_dict['성함'][i] != src2_dict['1. 이름 (예시:김보건)(*)'][idx]: print("Error:", dst_dict['성함'][i], "!=", src2_dict['1. 이름 (예시:김보건)(*)'][idx], phone)
    dst_dict['생년월일'][i] = src2_dict['4. 생년월일 (예시: 2000.01.01)(*)'][idx]
    dst_dict['1365 ID'][i] = src2_dict['6. 1365 아이디(*)'][idx]
    dst_dict['VMS ID'][i] = src2_dict['7. VMS 아이디(*)'][idx]
```

```python
dst = pd.DataFrame.from_dict(data=dst_dict, orient='columns')
dst.to_excel(root_path+"/n차_봉사자_수기명부.xlsx", index=False, engine="openpyxl", header=True) # xlsx 파일로 저장
dst
```
