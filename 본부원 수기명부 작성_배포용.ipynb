```python
import pandas as pd
import re
import numpy as np
from datetime import datetime

root_path = "/content/drive/MyDrive/Code/IHCO/" ## 경로 변경 요망
src = pd.read_excel(root_path+"dataset/설문조사_배포용.xlsx")
src = src.loc[:, ['1. 이름 (예시:김보건)(*)', '4. 생년월일 (예시: 2000.01.01)(*)', '5. 연락처(*)', '6. 1365 아이디(*)', "7. VMS 아이디(*)", "8. '보건의료통합봉사회' 의 소속을 밝혀주세요.(*)"]]
src = src[src["8. '보건의료통합봉사회' 의 소속을 밝혀주세요.(*)"]=="서울 응암 본부원입니다."].iloc[:, :-1]
src.columns = ['성함', '생년월일', '전화번호', '1365 ID', 'VMS ID'] # 컬럼명 변경
src.reset_index(inplace=True, drop=True) # index 재설정
```

```python
columns = ['순번', '활동일자', '활동시간', '성함', '생년월일', '전화번호', '1365 ID', 'VMS ID', '사진 및 동영상 촬영 동의', '개인정보 활용 동의', '체온', '서명']
for f in columns:
  if f not in src.columns: src[f] = np.NaN # 컬럼 생성
src['순번'] = src.index+1
date = input("n차 봉사 날짜 입력(예: 24.01.01): ")
src['활동일자'] = datetime.strptime(date,  '%y.%m.%d')
src['활동시간'] = '10:00~18:00'
src = src[columns]
src.to_excel(root_path+"dataset/본부원 전체 수기명부.xlsx", index=False, engine="openpyxl", header=True) # xlsx 파일로 저장
src
```
